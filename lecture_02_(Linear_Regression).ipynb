```python
df = pd.read_csv("area.csv")
df.head()
```

```python
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(df.area,df.price, color = 'r')
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']], df['price'])
```

```python
model.predict([[5000]])
```

```python
model.coef_*5000+model.intercept_
```
